## Experimental data if feasible

To prove smoking is harmful, a designed experiment can be helpful:

* Randomly assign individuals into two groups.
* Individuals in Group one are forced to smoke many packs of cigarettes.
* Group two are kept away from cigarettes.
* Keep all other aspects of individuals and lifestyle as similar as possible.
* record the lung cancer incident after 10 years.

- If the individuals in the smoking group have a higher cancer rate based on data from this experiment, it becomes harder to blame genes or other factors for it.

## What is the experimental design?


It specifies the conditions under which the experiment to
be carried out and what kind of data will be collected.

These specifications are designed to avoid interference of factors not of interest and enable the maximum amount of information on factors of interest.

This course introduces some classical and basic designs.
However, much of our attention is on the property of these
designs and subsequent data analysis strategies.

## Advantages of the designed experiment

Through this course, I hope to convince you that a well-designed experiment makes scientific investigation:

* more efficient: arrive at the conclusion at a low cost in money and time.
* more conclusive: isolate the factors you wish to study by removing the effect of other factors.
* easy to analyze: collect data with a simple structure to ease the task of data analysis.

## Features of the designed experiment

By designing the experiment, we make it

* comparative: two or more treatments are compared in a clinical trial.

* have well-defined experimental units: divide the experiment objects into individual units, assign one treatment to each unit.

   - examples: patients in clinical trials, eyes of a person, parts of a machine, plots in a field, materials for chemical reactions.

## Features of the designed experiment

* under control: the allocation of treatments to experimental units is decided within the experimental protocol.

* randomized: the set of units receiving specific treatment is pre-decided at random to limit the influence of non-anticipated factors.

* blocked: if experiment units can be divided into more uniform groups, ensure the comparisons are made under same conditions.

* carefully planned: of course.

## Types of experiments

1. Treatment comparisons. 

The main purpose is to compare
several treatments and select the best ones. 

Normally, it implies
that a product can be obtained in several different ways, and
we want to know which one is the best according to some standard/criterion.

## Types of experiments

2. Variable screening. 

The output is likely influenced by a number of factors. 

For instance, the chemical reaction is controlled by
temperature, pressure, concentration, duration, operator, and so on. 

Is it possible that only some of them are crucial and some of them 
can be dropped from consideration?

## Types of experiments

3. Response surface exploration. 

Suppose a few factors have been
identified to have a meaningful impact on the output. 

We may then search for a simple mathematical relationship between the values of these factors and the output.

Linear regression and response surfaces are often used.

## Types of experiments

4. System optimization. 

The purpose of most (statistical) experiments
is to find the best possible set of the input variables.

The output of an experiment can be analyzed to help us to
achieve this goal.

## Types of experiments

5. System robustness. 

Suppose the system is approximately 
optimized at two (or more) possible settings of the input 
variables. 

However, in mass production, it could be costly to 
control the input variables precisely.

The system deteriorates when the values of the input
variables deviate from these settings.

A setting is most robust if the system deteriorates least when the system wonders away from the optimal setting.

## A systematic approach 
 to the planning and implementation of the experiment.

The following five-step procedure is generally recommended.

* 1. State objectives. What do we hope to learn?

    The answer is usually from your future boss. It could be something we want to demonstrate and hope that the outcome will impress your future boss.


* 2. Determine the response.

    What do we plan to observe?


### systematic approach (continued)

* 3. Choose factors and levels.

    To study the effect of factors, two or more levels of each factor are needed. 

    - Factors may be quantitative and qualitative.  

    - How much fertilizer we use is a quantitative factor. 

    - What kind of fertilizer we use is a qualitative factor.

    Is it possible that only some of them are crucial and some of them can be dropped from consideration? Have we included all influential factors?


### systematic approach (continued)

* 4. Work out an experimental plan.

    The key consideration is to obtain the information efficiently. Poor design may capture little information that no analysis can rescue. 


* 5. Perform the experiment.

    Make sure we carry out the experiment as planned.
        - If practical situations arise such that we have to alter the plan, be sure to record it in detail.


### Terminologies of DoE

* The output variable in an experiment is also called the -response-. 

* The input variables are referred to as -factors-, with different -levels- that can be controlled or set by the experimenter.

* A -treatment- is a combination of factor levels. When there is only one factor, each level defines a treatment. 

* An -experimental unit- is a generic term that refers to a basic unit such as material, animal, person, plant, time period, or location, to which a treatment is applied. 

* The process of choosing a treatment, applying it to an experiment unit, and obtaining the response is called an  -experimental run- or a -trial-.

## A few lines of probability theory

* Random variables are real-valued functions on $\Omega$.

* e.g.: toss 10 coins leads to 1024 possible outcomes in terms of H and T.

* Let $X$ be the number of H in the outcome of 10 coins.

* This statement forces $X$ to translate every 1024 possible outcomes into a number.

* $X$ itself is not so random. The experiment of tossing 10 coins leads to random outcomes 
and therefore randomness in the observed value of $X$.

## A few lines of probability theory

-- random variable leads to distribution.

* There is exactly one outcome leads to $X=0$; 10 outcomes lead to $X=1$.

* We read $X=1$ as a statement that one of these 10 outcomes occurred.

* This interpretation leads to $P(X=1) = 10*(1/2)^{10}$

* Similarly, $P(X=k) = {n \choose k} p^k(1-p)^{n-k}$ in a general setting.

## A few lines of probability theory

-- random variable leads to distribution.

* Suppose $X$ is a random variable;

* and it can take values 0, 1, 2, $\ldots$ only;

* and it has specific values of $P(X=i)$ for all $i=0, 1, \ldots$.

Then we have completely specified its randomness.

## A few lines of probability theory

-- Distribution of random variables.

* Suppose $X$ is a random variable and we have specified the expression of $P(X \leq t)$ for any real number $t$.

* Knowing $P(X \leq t)$ implies we know everything about $X$ itself. This function is the cdf of $X$, often denoted as $F_X(t)$.

* We name the distribution of a random variable in terms of its form of $F_X(t) = P(X \leq t)$.

* For instance, when 
$$P(X \leq t) = \{ 1 - \exp(-t)\} I(t \geq 0),$$ 
we say $X$ has standard exponential distribution.

## A few lines of probability theory

-- Distribution of random variables.

* Sometimes, $F_X(t)$ can be derived from other functions.

* e.g.  $F_X(t) = \{ 1 - \exp(-t)\} I(t \geq 0)$ has a derivative function $f_X(t) = \exp(-t) I(t \geq 0)$.

* knowing $f_X(t) = \exp(-t) I(t \geq 0)$ is just as good as knowing $F_X(t)$.

* We call $f_X(t)$ the pdf of $X$. 

* Too often, we abuse notation $f(x)$ as if it stands for pdf of any distribution, any random variable.
